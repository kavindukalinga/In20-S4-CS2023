Comparing time taken for $ Recursive Quick Sort Algorithm$ and $Iterative Quick Sort Algorithm $.


```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as pyplot

# Manual data setup
arraylength = np.array([500,1000,5000,10000,20000])

srt1 = np.array([318000,1246000,28638000,60154000,125681000])
spline1=make_interp_spline(arraylength,srt1)
arraylength1=np.linspace(arraylength.min(),arraylength.max(),500)
srts1=spline1(arraylength1)

srt2 = np.array([108000,1310000,27367000,107218000,423098000])
spline2=make_interp_spline(arraylength,srt2)
arraylength2=np.linspace(arraylength.min(),arraylength.max(),500)
srts2=spline2(arraylength2)

# srt3=[654300,2522500,63499800,280970500,1362874100]
# spline3=make_interp_spline(arraylength,srt3)
# arraylength3=np.linspace(arraylength.min(),arraylength.max(),500)
# srts3=spline3(arraylength3)

# srt4=[494800,2266000,45090000,178934200,719721400]
# spline4=make_interp_spline(arraylength,srt4)
# arraylength4=np.linspace(arraylength.min(),arraylength.max(),500)
# srts4=spline4(arraylength4)

# line style formatting
pyplot.plot(arraylength1, srts1,label = 'Recursive-Quick Sort')
pyplot.plot(arraylength2, srts2,label = 'Itterative-Quick Sort')
# pyplot.plot(arraylength3, srts3,label = 'Optimized Bubble Sort')
# pyplot.plot(arraylength4, srts4,label = 'Selection Sort')
#pyplot.plot(year, subdomains, color = 'green',linestyle = 'dashed', linewidth = 3,label = 'Subdomains')

# layout configuration
pyplot.xlabel('Array Length')
pyplot.ylabel('Time ($1x10^{-1}$s)')

# Setting the title of the entire plot at the top.
pyplot.title('Time Taken for each sorting algorithm')

# A legend defines a region that describes the elements of the graph.
pyplot.legend()

# Save the chart file
pyplot.savefig('plot1.png', dpi=300)

# Print the chart
pyplot.show()
```
